# Craete Empty List 

```python
corpus=[]
```

```python
dataset['Review'].head()
```

```python
for i in range(0,1000):
    review=dataset['Review'][i]
    review=re.sub('[^a-zA-Z]',' ',review)
    review=review.lower()
    review=review.split()
    d=[]
    for word in review:
        if(word not in sw):
            d.append(ps.lemmatize(word))
    review=' '.join(d)
    corpus.append(review)
```

```python
corpus
```

# BOW- Bag of word

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=1500)
```

```python
x=cv.fit_transform(corpus)
```

```python
x=x.toarray()
```

```python
x
```

```python
x.shape
```

```python
cv.get_feature_names()
```

```python
y=dataset['Liked']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Building NN Model

```python
from keras.layers import Dense,Dropout
```

```python
from keras.models import Sequential
```

```python
model=Sequential()
```

# ADD input Layer

```python
model.add(Dense(input_dim=x_train.shape[1],kernel_initializer='random_uniform',activation='relu',units=200))
model.add(Dropout(0.8))
```

# ADD Hidden Layer

```python
model.add(Dense(kernel_initializer='he_uniform',activation='relu',units=130))
```

```python
model.add(Dense(kernel_initializer='he_uniform',activation='relu',units=130))
```

```python
model.add(Dense(kernel_initializer='he_uniform',activation='relu',units=130))
```

```python
model.add(Dense(kernel_initializer='he_uniform',activation='sigmoid',units=1))
```

# Compile the Model

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

# Train the NN Model

```python
model.fit(x_train,y_train,epochs=300,batch_size=32,validation_data=(x_test,y_test))
```

```python
y_pred=model.predict_classes(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
